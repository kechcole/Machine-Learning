# LINEAR REGRESSION.

THis is a ML algorith that is used to predict a single variable using another variable. The former is the dependent variable(Y) while the later which is used to predict is the independent variable(X). Linear algorithm fits a line that optimises errors such that there is extremely minor descripancies between the predicted value and actual output value. Using the "least square" method the algorithm generate a line that fits best in paired datasets. 

Given a dataset, X is the training feature while Y is the label while m and c are the training parameters , slope and y intercept. 




Thus, regression modeling is all about finding the values for the unknown parameters of the equation, i.e., values for 01 and 02 (weights), so that give X oen could easily predict the values of y.

The scikit-learn library in Python implements Linear Regression through the LinearRegression class. This class allows us to fit a linear model to a dataset, predict new values, and evaluate the model's performance.

To use the LinearRegression class, we first need to import it from sklearn.linear_model module. We can then create an instance of the class and call its fit method to train the model on a dataset. Finally, we can use the prediction method to generate predictions on new data.

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Import dataset
iris = sns.load_dataset('iris')

print(iris)
print(f'Structure : {iris.shape}')
print(f'Columns : {iris.columns}')
```

```python
# Subset the data to get only data needed , 3rd and 4th columns 
iris = iris[['petal_length', 'petal_width']]
print(iris)
```

```python
# View the relationship between variable
sns.lmplot(data=iris, x="petal_length", y="petal_width")
```

```python
# Separate data into dependent and independent variables
X = iris['petal_length']
y = iris['petal_width']

# Split the data into training(60%) and testing data(40%) ,pick random rows and sp
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state=23)

print(X_train)
```

```python
# Convert data to 2-D array 
X_train = np.array(X_train).reshape(-1, 1)
X_test = np.array(X_test).reshape(-1, 1)

print(X_train.shape)
```

```python
# Fit a model 

# Save model to variable 
regr = LinearRegression()

# Fit , search for line of best fit
regr.fit(X_train, y_train)
```

```python
# View parameters 

# Intercept 
c = regr.intercept_

# SLope of best fit line of the regression model 
m = regr.coef_

print(f'Y intercept : {c}')
print(f'Scale factor / regression coefficinet : {m}')
```

```python
# Predict y given training feature x by validating the model. 

# Beacause we are feeding the model with the data used to train it, we expect the performance extremely good. 
# This is called overfitting in ML, a model memorises the training data, this gives a false evaluation that our model 
# is extremely good because data was leaked into the test data.   
# The dange arising is that when the model is exposed to unseen test data, 


y_pred_train1 = regr.predict(X_train)

# Bestfit line
plt.scatter(X_train, y_train, color='green')
plt.plot(X_train, y_pred_train1, color='red')
plt.xlabel("petal_length")
plt.ylabel("petal_width")

```

```python
# Lets expose the data to never seen before data, test data. 

y_pred_test1 = regr.predict(X_test)

# Bestfit line
plt.scatter(X_test, y_test, color='green')
plt.plot(X_test, y_pred_test1, color='red')
plt.xlabel("petal_length")
plt.ylabel("petal_width")
```
